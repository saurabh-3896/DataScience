```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,3*np.pi,0.1)
ySin=np.sin(x)
yCos=np.cos(x)

#plt.plot(x,ySin)
#plt.plot(x,yCos)
plt.subplot(1,2,1)
plt.plot(x,ySin)
plt.xlabel('some x axis')
plt.ylabel('some y axis')
plt.title('Sine')

plt.subplot(1,2,2)
plt.plot(x,yCos)
plt.xlabel('some x axis')
plt.ylabel('some y axis')
plt.title('Cosine')

#plt.legend(['Sine','Cosine'])
plt.show()
```
